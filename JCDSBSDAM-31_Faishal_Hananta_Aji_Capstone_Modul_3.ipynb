## 1. Business Understanding

### 1.1. Problem Background

Pembatalan pemesanan/reservasi kamar hotel merupakan suatu permasalahan yang terus muncul dalam dunia perhotelan. Pemesanan atau reservasi kamar yang dibatalkan dapat menyebabkan pihak hotel mengalami kerugian pendapatan (profit) potensial. Untuk mengatasi permasalahan pembatalan reservasi kamar hotel, pihak hotel dapat melakukan beberapa strategi untuk mengurangi kerugian yang dialami oleh hotel, salah satunya berupa strategi **overbooking**.

Overbooking merupakan sebuah strategi dimana **suatu hotel menerima lebih banyak reservasi daripada jumlah kamar yang sebenarnya tersedia** [1]. Ketika hotel melakukan overbooking, **hotel tetap membuka pemesanan/reservasi sebuah kamar hotel walaupun kamar hotel tersebut sudah dipesan/direservasikan**. Hal ini dilakukan dengan anggapan bahwa beberapa pemesanan/reservasi yang sebelumnya telah diterima untuk kamar hotel tersebut akan dibatalkan. Dengan menerima beberapa pemesanan/reservasi kamar hotel untuk kamar yang sama di saat yang sama, hal ini memastikan bahwa bahkan ketika ada pemesanan/reservasi kamar hotel yang dibatalkan, masih ada pemesanan/reservasi yang diterima oleh hotel untuk kamar tersebut, sehingga pihak hotel tidak menerima kerugian pendapatan potensial dari pemesanan/registrasi yang dibatalkan. 

Walau strategi overbooking ini mungkin dapat menghindarkan pihak hotel dari kerugian potensial, **strategi ini dapat memunculkan risiko-risiko tertentu yang dapat memunculkan kerugian-kerugian lain pada pihak hotel**, terutama jika anggapan pihak hotel bahwa suatu pemesanan/reservasi kamar hotel akan dibatalkan ternyata tidak terjadi, sehingga **mengakibatkan terjadinya kasus dimana dua atau lebih pemesanan/reservasi kamar yang bertabrakan**. [2] 
- Kasus reservasi kamar yang bertabrakan dapat menyebabkan berkurangnya tingkat kebahagiaan/kepuasan pengunjung hotel, serta dapat mengakibatkan pengunjung hotel untuk enggan memesan kamar di hotel tersebut untuk kedepannya.
- Kasus reservasi kamar yang bertabrakan dapat menyebabkan penurunan reputasi hotel secara online di sosial media.
- Kasus reservasi kamar yang bertabrakan dapat menyebabkan munculnya beban yang berlebih pada staf hotel, terutama bagi staf front desk seperti resepsionis, yang langsung harus berhadapan dengan pengunjung yang tidak puas pada hotel.

Karena strategi overbooking dapat mengakibatkan munculnya kerugian pada pihak hotel karena adanya kasus reservasi yang bertabrakan, strategi ini tidak bisa secara serta merta digunakan. Agar pihak hotel tidak hanya bisa mengandalkan strategi overbooking, **diperlukan sebuah metode alternatif yang dapat memprediksi apakah seorang pengunjung akan membatalkan pesanan/reservasi kamarnya atau tidak**.

### 1.2. Problem Statement

Ketika seorang customer melakukan reservasi kamar hotel, akses kamar hotel tersebut akan disimpan oleh hotel untuk customer yang melakukan reservasi. Selama kurun waktu di antara waktu reservasi dilakukan hingga waktu customer yang melakukan reservasi datang untuk check-in, kamar tersebut tidak akan digunakan serta tidak akan bisa dipesan oleh calon-calon customer lainnya, sehingga pada kurun waktu tersebut, kamar-kamar tersebut tidak akan memberikan pendapatan apapun bagi pihak hotel. Apabila customer membatalkan pemesanan selama kurun waktu tersebut, pihak hotel akan mengalami kerugian pendapatan, karena kamar hotel tersebut akan disimpan secara sia-sia. 

Maka dari itu, **pihak hotel ingin dapat memprediksi apakah seorang calon customer yang ingin melakukan reservasi kamar nantinya akan membatalkan reservasi kamar tersebut dari dini hari.** Tidak hanya itu, **pihak hotel juga ingin melihat faktor-faktor yang paling mempengaruhi kemungkinan seorang customer untuk membatalkan reservasi kamar.**

### 1.5. Target and Evaluation Metric

Dari problem statement di atas, dapat diketahui bahwa model prediksi yang dibuat akan digunakan untuk memprediksi pembatalan reservasi kamar oleh customer. Maka dari itu, dapat diketahui bahwa target prediksi dari model yang dibuat adalah sebagai berikut:
- 1 (Positive) = **customer membatalkan reservasi kamar**
- 0 (Negative) = **customer tidak membatalkan reservasi kamar**

Kesalahan prediksi (error) yang dapat terjadi pada saat model melakukan prediksi juga dapat diketahui, dimana:
- Type 1 Error (False Positive): Customer diprediksikan akan membatalkan registrasi kamar, padahal customer sebenarnya tidak akan membatalkan registrasi kamar.
- Type 2 Error (False Negative): Customer diprediksikan tidak akan membatalkan registrasi kamar, namun setelah registrasi dilakukan, customer ternyata membatalkan registrasi kamar. 

Dengan asumsi bahwa pihak hotel akan menolak registrasi dari customer yang terprediksi akan membatalkan registrasi, kedua jenis kesalahan ini akan mengakibatkan terjadinya situasi dan konsekuensi sebagai berikut:
- Kasus false positive: **Customer ditolak registrasi kamarnya, karena dianggap akan membatalkan registrasi kamar, padahal customer tersebut tidak akan membatalkan registrasi kamar**. Kamar tersebut langsung bisa dipesan customer lain, karena tidak ada registrasi yang dilakukan.
    - Pihak hotel akan mengalami **kerugian potensial dari pemesanan tersebut**, dimana pihak hotel tidak akan dapat menerima keuntungan yang mungkin didapatkan dari pembayaran kamar hotel customer tersebut. 
- Kasus false negative: **Customer diterima registrasi kamarnya, namun registrasi tersebut kemudian dibatalkan oleh customer**. Kamar tersebut tidak akan dapat dipesan customer lain hingga customer membatalkan registrasi. Kurun waktu dari penerimaan registrasi kamar hotel hingga pembatalan registrasi dapat mencapai beberapa bulan.
    - Pihak hotel akan mengalami **kerugian potensial dari semua calon customer yang ingin memesan kamar tersebut selama kamar direservasi**, dimana pihak hotel tidak akan dapat menerima keuntungan yang mungkin didapatkan dari pembayaran kamar hotel dari tiap calon customer tersebut.

Dengan asumsi bahwa kamar hotel tersebut berharga 220 Euro, maka kerugian potensial yang diterima oleh pihak hotel untuk kasus false positive dan negative adalah:
- False positive = **220 Euro**.
- False negative = **220 Euro x jumlah potensial pemesanan kamar selama kamar direservasi**.

Seperti yang dapat dilihat di atas, kerugian potensial yang disebabkan oleh kasus false negative berpotensi **2 kali lipat atau lebih** dari kerugian potensial yang disebabkan oleh kasus false positive. Maka dari itu, model yang digunakan baiknya dapat **fokus dalam meminimalisir kemungkinan terjadinya false negative**. Walau begitu, **kasus false positive juga memberikan kerugian yang tidak bisa diabaikan**. Karena model yang ingin dibuat harus bisa meminimalisir false negative sembari tidak mengabaikan false positive, metrik evaluasi yang digunakan untuk mengevaluasi kinerja model adalah metrik **F1-score dengan pemberatan ke arah recall**, atau lebih tepatnya **F2-score**.

## 4. Feature Engineering

Sebelum model Machine Learning dapat dibuat, fitur yang digunakan perlu diolah terlebih dahulu. Hal ini dikarenakan bentuk fitur yang belum optimal dapat memperkeruh hasil prediksi model. Tidak hanya itu, fitur-fitur dengan bentuk tertentu tidak bisa digunakan dalam model Machine Learning sama sekali, sehingga perlu diubah ke bentuk yang dapat diolah. Hal ini dicapai melalui feature engineering dengan menggunakan:
- Scaling untuk mengubah rentang data agar setara di setiap fitur, serta 
- Encoding untuk mengubah data kategorikal ke dalam bentuk numerikal yang dapat diterima oleh model Machine Learning

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import (RobustScaler,  OneHotEncoder)
from category_encoders import BinaryEncoder
from imblearn.over_sampling import (RandomOverSampler, SMOTE)

from sklearn.impute import SimpleImputer

from sklearn.pipeline import Pipeline
from imblearn.pipeline import Pipeline as ImbPipeline

from sklearn.compose import ColumnTransformer
```

```python
df.head(1)
```

Sebelum data bisa digunakan untuk melatih model Machine Learning, data perlu dibagi menjadi target (variabel yang ingin ditebak) dan fitur (variabel yang digunakan untuk menebak nilai dari target) yang sudah didefinisikan sebelumnya di bagian "Business Understanding", dimana: 
- Kolom "is_canceled" merupakan target prediksi model, dan
- Kolom "country", "market_segment", "previous_cancellations", "booking_changes, "deposit_type", "days_in_waiting_list", "customer_type", "reserved_room_type", "required_car_parking_spaces", "total_of_special_requests" merupakan fitur yang digunakan untuk menebak target prediksi model.

Tidak hanya itu, data juga harus dibagi menjadi data latih (train) dan data uji (test), dimana data latih digunakan untuk melatih model untuk memprediksi target, sementara data uji digunakan untuk menguji model yang sudah dilatih. Data latih dan data uji tidak boleh sama, untuk menghindari terjadinya kebocoran data (data leak) yang akan membuat hasil prediksi menjadi tidak bisa diandalkan.

Agar tidak terjadi multikolinearitas, kolom "foreign" yang dibuat sebelumnya akan juga akan dihapus, karena nilai dari kolom ini diambil dari nilai pada kolom "country".

```python
# Menghapus kolom "foreign"
df.drop(columns="foreign", inplace=True)
```

Setelah kolom "foreign" dihilangkan, kolom yang dijadikan sebagai feature dan target dipisahkan, dimana:
- Target = kolom yang nilainya ingin diprediksi = `is_canceled`
- Feature = kolom yang nilainya digunakan untuk memprediksi target = semua kolom lainnya.

```python
# Mendefinisikan feature dan target
x = df.drop(columns="is_canceled")  # Fitur
y = df["is_canceled"]               # Target
```

```python
# Membelah data ke data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2,
                                                    random_state=42)

# Reset index
x_train.reset_index(drop=True, inplace=True)
y_train.reset_index(drop=True, inplace=True)
x_test.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)

print("Feature (Train) Shape: ", x_train.shape)
print("Target (Train) Shape: ", y_train.shape)
x_train.head()
```

```python
# Feature Engineering Pipeline

pipe_numeric = Pipeline(steps=[
    ("feature_imputation", SimpleImputer(strategy="median")),
    ("feature_scaling", RobustScaler())
])

pipe_nominal_categoric = Pipeline(steps=[
    ("feature_imputation", SimpleImputer(strategy="constant", fill_value="unknown")),
    ("feature_encoding", OneHotEncoder(drop="first", handle_unknown="ignore"))
])

pipe_binary_categoric = Pipeline(steps=[
    ("feature_imputation", SimpleImputer(strategy="constant", fill_value="unknown")),
    ("feature_encoding", BinaryEncoder())
])
```

```python
for col in df:
    print(f"Unique values in {col}: {df[col].nunique()}")
```

```python
# Transformer
transformer = ColumnTransformer(transformers=[
    ("pipeline_numerical", pipe_numeric, ["previous_cancellations", "booking_changes", "days_in_waiting_list", "total_of_special_requests", "required_car_parking_spaces"]),
    ("pipeline_nominal", pipe_nominal_categoric, ["deposit_type", "customer_type"]),
    ("pipeline_binary", pipe_binary_categoric, ["country", "market_segment", "reserved_room_type"])
], verbose_feature_names_out=False, remainder="passthrough")

transformer
```

```python
# Testing transformer
# Lihat apakah ada error
transformer.fit(x_train)
x_train_pre = transformer.transform(x_train)
x_test_pre = transformer.transform(x_test)
```

## 5. Modeling

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import (AdaBoostClassifier, 
                              GradientBoostingClassifier, 
                              RandomForestClassifier,
                              StackingClassifier, 
                              BaggingClassifier)
from xgboost.sklearn import XGBClassifier
from lightgbm import LGBMClassifier

from sklearn.metrics import (fbeta_score, 
                             make_scorer, 
                             classification_report, 
                             ConfusionMatrixDisplay)

from sklearn.model_selection import (cross_validate, StratifiedKFold)
```

```python
# List model
list_model = {
    "KNN": KNeighborsClassifier(),
    "LOGREG": LogisticRegression(random_state=42),
    "DT": DecisionTreeClassifier(random_state=42),
    "ADABOOST": AdaBoostClassifier(random_state=42),
    "GBS": GradientBoostingClassifier(random_state=42),
    "XGB": XGBClassifier(random_state=42), 
    "LGBM": LGBMClassifier(random_state=42), 
    "RF": RandomForestClassifier(random_state=42)
}
```

```python
# Evaluation metric
f2 = make_scorer(fbeta_score, beta=2)
```

```python
# Model Pipeline
clf = ImbPipeline(steps=[
    ("feature_engineering", transformer), 
    ("classifier", None)
])

clf_resampling = ImbPipeline(steps=[
    ("feature_engineering", transformer), 
    ("feature_resampling", None),
    ("classifier", None)
])
```

### 5.1. Model Benchmarking

```python
# Model benchmarking

benchmark_result = []

for model_name in list_model:
    clf.set_params(classifier=list_model[model_name])
    cv_result = cross_validate(estimator=clf, 
                               X=x_train, y=y_train,
                               scoring=f2, 
                               cv=StratifiedKFold(n_splits=5),
                               return_train_score=True)
    benchmark_result.append({
        "model_name": model_name, 
        "mean_train_score": np.mean(cv_result["train_score"]), 
        "std_train_score": np.std(cv_result["train_score"]),
        "mean_test_score": np.mean(cv_result["test_score"]), 
        "std_test_score": np.std(cv_result["test_score"])
    })
```

```python
benchmark_result = pd.DataFrame(benchmark_result)
benchmark_result.sort_values(by="mean_test_score", ascending=False, inplace=True)
benchmark_result
```

```python
list_resampler = {
    "ROS": RandomOverSampler(random_state=42),
    "SMOTE": SMOTE(random_state=42)
}

benchmark_with_resampling = []

for model_name in list_model:
    for resampler in list_resampler:
        clf_resampling.set_params(classifier=list_model[model_name], 
                                  feature_resampling=list_resampler[resampler])
        cv_result = cross_validate(estimator=clf_resampling, 
                                   X=x, y=y, 
                                   scoring=f2, 
                                   cv = StratifiedKFold(n_splits=5),
                                   return_train_score=True)
        benchmark_with_resampling.append({
        "model_name": model_name,
        "resampler": resampler, 
        "mean_train_score": np.mean(cv_result["train_score"]), 
        "std_train_score": np.std(cv_result["train_score"]),
        "mean_test_score": np.mean(cv_result["test_score"]), 
        "std_test_score": np.std(cv_result["test_score"])
        })
```

```python
# Benchmark result (with resampling)
benchmark_with_resampling = pd.DataFrame(benchmark_with_resampling)
benchmark_with_resampling.sort_values(["mean_test_score", "mean_train_score"], ascending=False)
```

Karena class dari data sudah balanced (1/3 banding 2/3), maka model yang digunakan merupakan model TANPA resampling, walaupun hasil model dengan resampling lebih baik sedikit.

Setelah benchmarking, dapat terlihat bahwa model yang menghasilkan training score dan testing score terbaik merupakan model **KNN dan Random Forest**.

    - Apabila model ingin digunakan untuk konteks hotel pada tempat lain, training model bisa diulang dengan data hotel baru agar hasil prediksi model lebih relevan dengan kondisi di tempat baru.
